 # <center> Principe de la dichotomie </center>
<div class="keyword-content">Nous construisons deux suites de r√©els $(a_n)_n$ et $(b_n)_n$ adjacentes telles que pour tout entier naturel $n$, $a_n\le \alpha \le b_n$. La construction de ces deux suites se fait par l&rsquo;interm√©diaire d&rsquo;une suite d&rsquo;intervalles $I_n=[a_n;b_n]$.Pour cela, nous proc√©dons par r√©currence&nbsp;:<ul><li>Posons $a_0=a$ et $b_0=b$ donc $I_0=[a;b]$.</li><li>Supposons qu&rsquo;il existe deux entiers $a_n$ et $b_n$ tels que $a_n\le \alpha \le b_n$ et $I_n=[a_n;b_n]$.Consid√©rons le milieu $\dfrac{a_n+b_n}{2}$ de l&rsquo;intervalle $I_n$. Deux cas se pr√©sentent&nbsp;:<ol><li>Si $f(a_n)f\left(\dfrac{a_n+b_n}{2}\right)\le 0$, alors $\alpha \in \left[a_n;\dfrac{a_n+b_n}{2}\right]$.Dans ce cas, nous posons $a_n+1=a_n$ et $b_n+1=\dfrac{a_n+b_n}{2}$.</li><li>Sinon $\alpha\in\left[\dfrac{a_n+b_n}{2};b_n\right]$.Nous posons alors $a_n+1=\dfrac{a_n+b_n}{2}$ et $b_n+1=b_n$.</li></ol>A chaque √©tape, nous prenons une moiti√© de l&rsquo;intervalle pr√©c√©dent, d&rsquo;o√π le nom de cette m√©thode puisque dichotomie signifie partage en deux parties.On montre facilement que les suites $(a_n)_n$ et $(b_n)_n$ sont adjacentes.D&rsquo;autre part&nbsp;:\[\text{Pour tout entier naturel $n$,}\ a_n\le \alpha \le b_n.\]Nous en d√©duisons&nbsp;:\[\lim a_n=\lim b_n=\alpha.\]Enfin&nbsp;:\[\text{Pour tout entier naturel $n$,}\ b_n-a_n=\frac{1}{2^n}(b-a).\]</li><li>Nous avons donc&nbsp;:\[\text{Pour tout entier naturel $n$,}\ 0\le\alpha -a_n\le\frac{1}{2^n}(b-a).\]Cette derni√®re relation permet de dire que le r√©el $a_n$ est une valeur approch√©e de $\alpha$ √† la pr√©cision $\dfrac{1}{2^n}(b-a)$.</li></ul>Cette m√©thode fournit un algorithme de calcul de r√©solution d&rsquo;une √©quation de la forme ¬´ $y=f(x)$ ¬ª.</div></div></p><p><br/>
